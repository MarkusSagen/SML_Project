```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA
from sklearn import preprocessing

songs_to_classify = pd.read_csv('songs_to_classify.csv')
```

```python
cont_feature_names = ['acousticness', 
                      'danceability' , 
                      'duration' , 
                      'energy' , 
                      'instrumentalness' , 
                      'liveness' , 
                      'loudness', 
                      'speechiness', 
                      'tempo', 
                      'valence']

disc_feature_names = ['mode', 'key', 'time_signature']
```

```python
# make cont part of dataset into numpy array  in order to scale 
cont_set_scaled_array = preprocessing.scale(songs_to_classify[cont_feature_names].values)
```

```python
#  Make a data frame of new scaled version of the cont. 
scaled_cont_songs = pd.DataFrame(cont_set_scaled_array, columns= cont_feature_names)
```

```python
# Add catagorical data back to the scaled dataset
scaled_cont_songs[disc_feature_names] = songs_to_classify[disc_feature_names]
```

```python
songs_scaled = scaled_cont_songs
```
