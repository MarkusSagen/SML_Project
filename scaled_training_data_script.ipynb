```python
# This notebok scales the training data produces a scaled version in csv version. 
# it also generates a new training set, features with near zero correlation to the label are simply dropped. 


# Import modules

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA
from sklearn import preprocessing

songs = pd.read_csv('training_data.csv')

# Dataframe for the sets of liked and disliked songs respectivly.
# python code form: df.loc[df['column_name'] == some_value]
liked_songs = songs.loc[(songs['label'] == 1)]
disliked_songs = songs.loc[(songs['label'] == 0)]



# names of discreate and cont fetaures. 
"""
'acousticness'
'danceability' 
'duration' 
'energy' 
'instrumentalness' 
'key' 
'liveness' 
'loudness'
'mode'
'speechiness' 
'tempo'
'time_signature'
'valence' 
'label'
"""
cont_feature_names = ['acousticness', 
                      'danceability' , 
                      'duration' , 
                      'energy' , 
                      'instrumentalness' , 
                      'liveness' , 
                      'loudness', 
                      'speechiness', 
                      'tempo', 
                      'valence']

disc_feature_names = ['mode', 'key', 'time_signature', 'label']
```

```python
# make cont part of dataset into numpy array  in order to scale 
cont_set_scaled_array = preprocessing.scale(songs[cont_feature_names].values)
```

```python
#  Make a data frame of new scaled version of the cont. 
scaled_cont_songs = pd.DataFrame(cont_set_scaled_array, columns= cont_feature_names)
```

```python
# Add catagorical data back to the scaled dataset
scaled_cont_songs[disc_feature_names] = songs[disc_feature_names]
```

```python
songs_scaled = scaled_cont_songs
```
